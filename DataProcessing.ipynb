```python
l1['Interest_Rate'] = l1['Interest_Rate'].str.replace('%', '').astype(float)
l1['Loan_Length'] = l1['Loan_Length'].str.replace(' months', '').astype(float)
l1['Debt_To_Income_Ratio'] = l1['Debt_To_Income_Ratio'].str.replace('%', '').astype(float)
```

```python
def clean_employment_length(length):
    if pd.isna(length):
        return np.nan
    length = str(length).lower()
    if 'year' in length:
        if '< 1' in length:
            return 0.0
        elif '10+' in length:
            return 10.0
        else:
            return float(length.replace(' years', '').replace(' year', ''))
    return np.nan

l1['Employment_Length'] = l1['Employment_Length'].apply(clean_employment_length).astype(float)
```
